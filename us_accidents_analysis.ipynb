### 2. Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde= False, norm_hist= True)
```

1. A high percentage of accidents occur between 2 P.M and 6 P.M
2. There is a spike during 6 A.M to 9 A.M indicating that the reason might be due to people leaving for work.

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde= False, norm_hist= True)
```

Is the distribution of accident by hour the same on weekends as on weekdays?

```python
Sunday_Start_Time = df.Start_Time[df.Start_Time.dt.dayofweek == 6 ]
sns.distplot(Sunday_Start_Time.dt.hour, bins=24, kde= False, norm_hist= True)
```

```python
Monday_Start_Time = df.Start_Time[df.Start_Time.dt.dayofweek == 0 ]
sns.distplot(Monday_Start_Time.dt.hour, bins=24, kde= False, norm_hist= True)
```

1. During Weekends (Saturday and Sunday) there is a spike in accidents during midnight to early hours in the morning than during weekdays where the number of incidences is low during that timeframe.
2. Despite a weekday or weekend, high number of accidents occur during 3 P.M to 6 P.M window.

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
sns.distplot(df_2018.Start_Time.dt.month, bins=12, kde= False, norm_hist= True)
```

1. Data is missing for the first few months of 2016
2. There also seems to be some data missing for the years 2019, 2020 and 2021
3. We don't have the data on where the data was sourced from so as to check for irregularities in the dataset.
